```python
import numpy as np
import pandas as pd

data = pd.read_csv('/content/food_consumption.csv')
countries = list(data.iloc[:, 0])
x = data.iloc[:, 1:].to_numpy()

print(countries)
print(x.shape)
```

```python
from sklearn.decomposition import PCA

model = PCA(n_components=2)
model.fit(x)
new_x = model.transform(x)

print('x', x.shape)
print('new_x', new_x.shape)
```

```python
import matplotlib.pyplot as plt

plt.figure(figsize=(3, 3))
plt.scatter(new_x[:, 0], new_x[:, 1])
for i, country in enumerate(countries):
  plt.annotate(country, (new_x[i, 0], new_x[i, 1]))
plt.show()
```
