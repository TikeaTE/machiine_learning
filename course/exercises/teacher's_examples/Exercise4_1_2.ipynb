```python
import numpy as np
import pandas as pd
import torch

data = pd.read_csv('/content/data_3_2.csv').to_numpy()
x = data[:, :-1]
y = data[:, -1]

#feature scaling
x = (x - np.mean(x, axis=0)) / np.std(x, axis=0)

x = torch.tensor(x, dtype=torch.float32) #tensor(m, 2)
y = torch.tensor(y, dtype=torch.long) #tensor(m), must be of type long to work with CrossEntropyLoss
```

```python
import torch.nn as nn

class MyNN(nn.Module):
  def __init__(self):
    super().__init__()
    self.hid1_layer = nn.Linear(2, 5)
    self.hid2_layer = nn.Linear(5, 5)
    self.out_layer = nn.Linear(5, 4)

  def forward(self, x): #tensor(m, 2)
    output = self.hid1_layer(x) #tensor(m, 5)
    output = torch.sigmoid(output)
    output = self.hid2_layer(output) #tensor(m, 5)
    output = torch.sigmoid(output)
    output = self.out_layer(output) #tensor(m, 4)
    #output = torch.sigmoid(output), no need. Built-in act func in CrossEntropyLoss
    return output
```

```python
#Training
import torch.optim as optim

model = MyNN()
cost_func = nn.CrossEntropyLoss()
opt = optim.Adam(model.parameters(), lr=0.1)

for i in range(50):
  output = model(x)
  cost = cost_func(output, y)
  print('i: %d, Cost: %f' % (i, cost.item()))
  cost.backward()
  opt.step()
  opt.zero_grad()
```

```python
#Evaluating

output = model(x) #tensor(m, 4)
predict = torch.argmax(output, dim=1) #tensor(m)
acc = (predict == y).float().mean()
print('Accuracy: %f' % acc.item())
```
