```python
import numpy as np
import pandas as pd

data = pd.read_csv('/content/data_3_1_1.csv').to_numpy()
x = data[:, :-1]
y = data[:, -1]

print('x', x.shape)
print('y', y.shape)
```

```python
#feature scaling
x_scaled = (x-x.mean(axis=0))/x.std(axis=0)
```

```python
#Build and train the model z = a*x1+b*x2+c

import torch
import torch.optim as optim
import torch.nn as nn

tx = torch.tensor(x_scaled, dtype=torch.float32) #tensor(100, 2)
ty = torch.tensor(y, dtype=torch.float32) #tensor(100)

model = nn.Linear(2, 1)
opt = optim.SGD(model.parameters(), lr=1)
cost_func = nn.BCELoss()

for i in range(100):
  tz = model(tx)
  tz = torch.sigmoid(tz) #tensor(100, 1)
  tz = tz.flatten() #tensor(100)

  J = cost_func(tz, ty)

  print('i: %d, Cost: %f' % (i, J.item()))

  J.backward()

  opt.step() #update the parameters
  opt.zero_grad() #reset the gradient of each parameter
```

```python
#Evaluate the model

tz = model(tx)
tz = torch.sigmoid(tz)
tz = tz.flatten()
predict = (tz>=0.5).float()

acc_rate = (predict == ty).float().mean()
print('Accuracy: %f' % acc_rate)
```
