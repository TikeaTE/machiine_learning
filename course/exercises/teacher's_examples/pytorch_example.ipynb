```python
import torch
import numpy as np

a = 10
ta = torch.tensor(a, dtype=torch.float32)
print('a', a)
print('ta', ta)

b = [1, 2, 3]
tb = torch.tensor(b, dtype=torch.float32)
print('b', b)
print('tb', tb)

c = np.random.randn(2, 3)
tc = torch.tensor(c, dtype=torch.float32)
print('c', c)
print('tc', tc)
```

```python
a = torch.tensor(10, dtype=torch.float32) #must be of type float
a.requires_grad = True
b = torch.tensor(20, requires_grad=True, dtype=torch.float32)
x = torch.tensor(100)
y = torch.tensor(200)

z = a*x+b
J = (z-y)**2

J.backward()

print('a.grad', a.grad)
print('b.grad', b.grad)
```
