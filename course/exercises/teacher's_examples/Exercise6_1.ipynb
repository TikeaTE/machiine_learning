```python
import numpy as np
import pandas as pd

x = pd.read_csv('/content/data_6_1.csv').to_numpy()
print(x.shape)
```

```python
from sklearn.cluster import KMeans

model = KMeans(n_clusters=5)
model.fit(x)
```

```python
print(model.cluster_centers_)
print(model.labels_[:10])
```

```python
#Elbow Method
import matplotlib.pyplot as plt

all_k = np.arange(start=2, stop=11)
costs = []
for k in all_k:
  model = KMeans(n_clusters=k)
  model.fit(x)
  costs.append(model.inertia_)

plt.figure(figsize=(4, 4))
plt.scatter(all_k, costs)
plt.show()
```
