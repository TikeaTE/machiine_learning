```python
import pandas as pd
import numpy as np

data = pd.read_csv('/content/land_price_1.csv').to_numpy()
print(data.shape)

x1 = data[:, 0]
x2 = data[:, 1]
y = data[:, 2]
```

```python
#Feature Scaling

x1_scaled = (x1-x1.mean())/x1.std()
x2_scaled = (x2-x2.mean())/x2.std()
```

```python
#Build and train the model z = a*x1+b*x2+c

import torch

tx1 = torch.tensor(x1_scaled, dtype=torch.float32)
tx2 = torch.tensor(x2_scaled, dtype=torch.float32)
ty = torch.tensor(y, dtype=torch.float32)

#Initialize all parameters of the model
a = torch.tensor(0, dtype=torch.float32, requires_grad=True)
b = torch.tensor(0, dtype=torch.float32, requires_grad=True)
c = torch.tensor(0, dtype=torch.float32, requires_grad=True)

alpha = 0.1
for i in range(50):
  tz = a*tx1+b*tx2+c #tensor(30)
  J = ((tz-ty)**2).mean()

  print('i: %d, Cost: %f' % (i, J.item()))

  J.backward()

  with torch.no_grad():
    a += -alpha*a.grad
    b += -alpha*b.grad
    c += -alpha*c.grad
    #reset the gradients
    a.grad.zero_()
    b.grad.zero_()
    c.grad.zero_()
```

```python
#Evaluate the model

tz = a*tx1+b*tx2+c

for zi, yi in zip(tz, ty):
  print('Predicted: %f, Actual: %f' % (zi, yi))
```
