```python
import cv2
import numpy as np

img = cv2.imread('/content/Beautiful-Flowers.jpg', cv2.IMREAD_COLOR_RGB)

print(img.shape)
```

```python
x = img.reshape((-1, 3))
print(x.shape)
```

```python
from sklearn.cluster import KMeans

model = KMeans(n_clusters=10)
model.fit(x)
```
