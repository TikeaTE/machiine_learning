```python
import numpy as np
import cv2
import glob

path = '/content/drive/MyDrive/CODING/dataset/tiny_mnist'

imgs = []
lbls = []
for i in range(3):
  img_files = glob.glob(path + '/*' + str(i) + '/*.png')
  for f in img_files:
    img = cv2.imread(f, cv2.IMREAD_GRAYSCALE) #np(28, 28)
    imgs.append(img)
    lbls.append(i)

imgs = np.array(imgs)
lbls = np.array(lbls)

print('imgs', imgs.shape)
print('lbls', lbls.shape)
```

```python
import torch

x = torch.tensor(imgs, dtype=torch.float32)/255.0 #tensor(m, 28, 28)
x = x.reshape(-1, 28*28) #tensor(m, 784)
y = torch.tensor(lbls, dtype=torch.long)
```
