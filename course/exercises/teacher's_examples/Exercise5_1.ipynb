```python
import numpy as np
import pandas as pd

data = pd.read_csv('/content/data_3_1_2.csv').to_numpy()
x = data[:, :-1]
y = data[:, -1]

#feature scaleing
x = (x - np.mean(x, axis=0)) / np.std(x, axis=0)
```

```python
m_train = int(0.6 * x.shape[0])
x_train = x[:m_train]
y_train = y[:m_train]
x_test = x[m_train:]
y_test = y[m_train:]
```

```python
from sklearn.linear_model import LogisticRegression
model = LogisticRegression()
model.fit(x_train, y_train)
```

```python
#Evaluation

def compute_PRF(z, y):
  TP = np.sum((z == 1) & (y == 1))
  FN = np.sum((z == 0) & (y == 1))
  FP = np.sum((z == 1) & (y == 0))
  precision = TP / (TP + FP)
  recall = TP / (TP + FN)
  f_score = 2 * precision * recall / (precision + recall)
  return precision, recall, f_score

precision, recall, f_score = compute_PRF(model.predict(x_test), y_test)

print(f'Precision: {precision}')
print(f'Recall: {recall}')
print(f'F-score: {f_score}')
```

```python
import torch
import torch.nn as nn

class MyModel(nn.Module):
  def __init__(self):
    super().__init__()
    self.layers = nn.Sequential(
        nn.Linear(2, 5),
        nn.ReLU(),
        nn.Linear(5, 3),
        nn.ReLU(),
        nn.Linear(3, 1),
        nn.Sigmoid()
    )

  def forward(self, x):
    output = self.layers(x)
    return output.flatten()
```

```python
import torch.optim as optim

tx = torch.tensor(x_train, dtype=torch.float32)
ty = torch.tensor(y_train, dtype=torch.float32)

model = MyModel()
opt = optim.Adam(model.parameters(), lr=0.1)
cost_func = nn.BCELoss()

for i in range(50):
  output = model(tx)
  loss = cost_func(output, ty)
  print(f'i: {i}, Loss: {loss.item()}')
  loss.backward()
  opt.step()
  opt.zero_grad()
```

```python
#Evaluation

tx_test = torch.tensor(x_test, dtype=torch.float32)

with torch.no_grad():
  output = model(tx_test)
  output = (output > 0.5).int().numpy()
  precision, recall, f_score = compute_PRF(output, y_test)

  print(f'Precision: {precision}')
  print(f'Recall: {recall}')
  print(f'F-score: {f_score}')
```
