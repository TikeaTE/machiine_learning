```python
#Data
import numpy as np

x = np.array([90, 130, 210, 300, 350, 420, 480, 530, 640, 710], dtype=float)
y = np.array([7.1, 10.9, 19.2, 28, 32.8, 39.9, 46.1, 51, 62.2, 68.9], dtype=float)
```

```python
#Training the Linear Regression Model using Gradient Descent
#h(x) = z = a*x+b

#Initialize the parameters
a = 0
b = 0

alpha = 0.000001

for i in range(50):
  z = a*x+b
  J = ((z-y)**2).mean()

  print('i: %d, Cost: %f, a: %f, b: %f' % (i, J, a, b))

  grad_a = ((z-y)*x).mean()*2
  grad_b = (z-y).mean()*2

  a += -alpha*grad_a
  b += -alpha*grad_b
```

```python
#Evaluating the model

z = a*x+b

for zi, yi in zip(z, y):
  print('Predicted: %.2f, Actual: %.2f' % (zi, yi))
```
