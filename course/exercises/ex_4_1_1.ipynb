```python
import numpy as np
import pandas as pd
import torch

data = pd.read_csv('/home/tikea/ecam_ws/ecam_courses/machine_learning/data_3_1_1.csv').to_numpy()
x = data[:, :-1]
y = data[:, -1]

#feature scaling
x = (x - np.mean(x, axis=0)) / np.std(x, axis=0)

x = torch.tensor(x, dtype=torch.float32) #tensor(m, 2)
y = torch.tensor(y, dtype=torch.float32) #tensor(m)
```

```python
import torch.nn as nn

class MyNN(nn.Module):
  def __init__(self):
    super().__init__()
    self.hid1_layer = nn.Linear(2, 5)
    self.hid2_layer = nn.Linear(5, 5)
    self.out_layer = nn.Linear(5, 1)

  def forward(self, x): #tensor(m, 2)
    output = self.hid1_layer(x) #tensor(m, 5)
    output = torch.sigmoid(output)
    output = self.hid2_layer(output) #tensor(m, 5)
    output = torch.sigmoid(output)
    output = self.out_layer(output) #tensor(m, 1)
    output = torch.sigmoid(output)
    output = output.squeeze(dim=1) #tensor(m)
    return output
```

```python
#Training
import torch.optim as optim

model = MyNN()
cost_func = nn.BCELoss()
opt = optim.Adam(model.parameters(), lr=0.1)

for i in range(50):
  output = model(x)
  cost = cost_func(output, y)
  print('i: %d, Cost: %f' % (i, cost.item()))
  cost.backward()
  opt.step()
  opt.zero_grad()
```

```python
#Evaluating

output = model(x)
output = (output > 0.5).float()
acc = (output == y).float().mean()
print('Accuracy: %f' % acc.item())
```
